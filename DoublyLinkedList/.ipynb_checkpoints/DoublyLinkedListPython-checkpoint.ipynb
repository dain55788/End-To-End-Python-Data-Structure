## Practice Doubly LinkedList in Python

```python
# Initialization step (create a node first):
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
```

```python
# Create a DoublyLinkedList Class that include its methods:
class DoublyLinkedList:
    def __init__(self):
        self.head = None
    
    #Check if the linkedlist is Empty:
    def is_empty(self):
        if self.head is None:
            return True
        return False
    
    # Return the length of a linkedlist:
    def take_length(self):
        length = 0
        temp = self.head
        while temp is not None:
            temp = temp.next
            length += 1
        return length
    
    # Search linearly a value in a linkedlist:
    def search_val(self, data):
        temp = self.head
        isFound = False
        while temp is not None:
            if temp.data == data:
                isFound = True
                break
            temp = temp.next
        return isFound
    
    #Insert at begin method:
    def insert_at_begin(self, data):
        new_node = Node(data)
        if self.is_empty():
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
        
    # Insert at ending of the linkedlist:
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.is_empty():
            self.head = new_node
        temp = self.head
        while temp is not None:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp
    
    # Insert a node before a given node
    def insert_before_node(node, data):
        if node is None:
            print("Error: The given node is None")
            return

        new_node = Node(data)
        new_node.prev = node.prev
        new_node.next = node

        if node.prev:
            node.prev.next = new_node

        node.prev = new_node

    
    # Insert after a specific node:
    def insert_after_node(node, element, value):  
        
        temp = self.head
        while temp is not None:
            if temp.data == element:
                break
            temp = temp.next
        if temp is None:
            print("{} is not present in the linkedlist, {} cannot be inserted into the list.".format(element, value))
        else: 
            new_node = Node(value)
            new_node.next = temp.next
            new_node.prev = temp
            temp.next.prev = new_node
            temp.next = new_node
    
    #Print elements in a linkedlist:
    def print_list(self):
        temp = self.head
        while temp is not None:
            temp = temp.next
            print(temp.data)
            
    #Update a specific value in linkedlist:
    def update_value(self, old_val, new_val):
        temp = self.head
        isUpdated = False
        while temp is not None:
            if temp.data == old_value:
                temp.data = new_value
                isUpdated = True
            temp = temp.next
        if isUpdated:
            print("Successfully update value!")
        else:
            print("Cannot find value!")
            
    #Delete the beginning of linkedlist:
    def del_beginning(self):
        if self.head is None:
            print("Cannot delete because your linkedlist is Null!")
        elif self.head.next is None: #there is one element in linkedlist:
            self.head = None
        else: #there are 2 or more elements the linkedlist:
            self.head = self.head.next
            self.head.prev = None 
    
    #Delete the last element of the linkedlist:
    def del_last(self):
        if self.head is None:
            print("Cannot delete because your linkedlist is Null")
        elif self.head.next is None:
            self.head = None
        else:
            temp = self.head
            while temp is not None:
                temp = temp.next
            temp.prev.next = None
            temp.prev = None
    
    #Delete a specific value in the linkedlist:
    def delete(self, value):
        if self.is_empty():
            print("Empty linkedlist, cannot delete!!")
            return
        elif self.head.next is None:
            self.head = None
        else:
            temp = temp.head
            while temp is not None:
                if temp.data == value:
                    break
                temp = temp.next
            if temp is None:
                print("Element not present in linked list. Cannot delete element.")
                return
            elif temp.next is None:
                self.del_last()
            else:
                temp.next = temp.previous.next
                temp.next.previous = temp.previous
                temp.next = None
                temp.previous = None
```
